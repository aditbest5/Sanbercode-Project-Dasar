Dapat dilihat banyak kolom yang terdapat outliernya. Untuk menghilangkan outlier dapat menggantinya dengan nilai IQR, mean, median, dan modus. Disini kita akan mengganti nilai outlier dengan nilai IQR

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR=Q3-Q1
data=data[~((data<(Q1-1.5*IQR))|(data>(Q3+1.5*IQR))).any(axis=1)]
 
# Cek ukuran dataset setelah kita drop outliers
data.shape
```

## **Standarisasi menggunakan StandardScaler**

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

# Standarisasi menggunakan StandardScaler
scaler = preprocessing.StandardScaler().fit(new_data)
new_data = scaler.transform(new_data)
new_data
```

## **Melakukan Clustering menggunakan K-Means dan Elbow Methods**

Kita akan melakukan clustering menggunakan K-Means dan mencari nilai optimum cluster nya menggunakan Elbow Methods. 

Berikut hasilnya :

```python
#Clustering dengan K-Means
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

kmeans1 = KMeans(n_clusters=2, random_state=42).fit(new_data)
labels1 = kmeans1.labels_

data_std = pd.DataFrame(data = new_data, columns = ['GDPperkapita', 'Pendapatan'])
data_std['label_kmeans1'] = labels1 

#Memvisualisasikan Hasil Clustering menggunakan scatterplot
plt.figure(figsize=(12,8))
plt.scatter(data_std.GDPperkapita[data_std.label_kmeans1 == 0], data_std['Pendapatan'][data_std.label_kmeans1 == 0], c='blue', s=100, edgecolor='green',linestyle='-')
plt.scatter(data_std.GDPperkapita[data_std.label_kmeans1 == 1], data_std['Pendapatan'][data_std.label_kmeans1 == 1], c='red', s=100, edgecolor='green',linestyle='-')

centers = kmeans1.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=500);
plt.xlabel('GDPperkapita')
plt.ylabel('Pendapatan')
plt.show()
```

```python
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters= i, init='k-means++', random_state = 42 )
  kmeans.fit(data_std)
  wcss.append(kmeans.inertia_)

plt.figure(figsize= (12, 8))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
