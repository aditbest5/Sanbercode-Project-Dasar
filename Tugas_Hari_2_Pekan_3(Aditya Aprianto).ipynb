Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, cos, color='green', linestyle='solid', label='Cosx') # menggunakan nama dari warnanya
plt.plot(x, cos3, color='red', linestyle='dashed', label='3 Cosx') # menggunakan format rgbcmyk
plt.plot(x, sin, color='blue', linestyle='dashdot', label='Sin(x-90)') # menggunakan format grayscale (0 1)
plt.plot(x, sin2, 'yx', label='3 Sinx') # menggunakan format RRGGBB, value from 00 to FF
plt.legend(loc='center')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Grafik Cos Sin')
```
