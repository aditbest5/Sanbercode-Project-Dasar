<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 8), sharex=True, sharey=True, squeeze=False)
ax[0][0].plot(x, tan, label='tanx', color='blue')
ax[0][0].set_title('Grafik tanx')
ax[0][1].plot(x, sin, label='sinx', color='red')
ax[0][1].set_title('Grafik sinx')
ax[0][2].plot(x, cos, label='cosx', color='green')
ax[0][2].set_title('Grafik cosx')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, y)
ax.set_title('Grafik CosX')

ins1 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins1.plot(x, y)
ins1.plot(x, y5)
ins1.set_title('Grafik CosX^5')

ins2 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins2.plot(x, y)
ins2.plot(x, y4, color='black')
ins2.set_title('Grafik CosX^4')

ins3 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3.plot(x, y)
ins3.plot(x, y2, color='red')
ins3.set_title('Grafik CosX^2')

ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins2.plot(x, y)
ins2.plot(x, y3, color='green')
ins2.set_title('Grafik CosX^3')

```
